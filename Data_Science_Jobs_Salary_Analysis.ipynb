```python
# Define the data
df = pd.read_csv('ds_salaries.csv')
```

```python
# Top job titles
plt.figure(figsize=(8,4))
df['job_title'].value_counts().head(10).plot(kind='barh', color='skyblue')
plt.title('Top 10 Job Titles')
plt.xlabel('Number of Jobs')
plt.ylabel('Job Title')
plt.show()
```

#***Top 10 Job Titles***
The bar chart below shows the most common data science job titles in this dataset.

**Insight:**  
- The most frequent titles include Data Scientist, Data Engineer, and Machine Learning Engineer.
- These roles dominate the data science field and reflect current hiring demand.

```python
# Average salary by job title
avg_salary = df.groupby('job_title')['salary_in_usd'].mean().sort_values(ascending=False).head(10)
plt.figure(figsize=(8,4))
avg_salary.plot(kind='bar', color='salmon')
plt.title('Average Salary (USD) by Job Title')
plt.ylabel('Salary in USD')
plt.show()
```

***# Average Salary by Job Title***

This chart displays the top 10 job titles with the highest average salary in USD.

**Insight:**  
- Senior roles like Principal Data Scientist and Machine Learning Manager have the highest average salaries.
- Some specialized roles (e.g., Data Science Manager) command significantly higher pay.

```python
# Remote vs On-site jobs
plt.figure(figsize=(5,5))
df['remote_ratio'].value_counts().plot(kind='pie', autopct='%1.1f%%', startangle=90)
plt.title('Remote vs On-Site Jobs')
plt.ylabel('')
plt.show()
```

# ***Remote vs On-Site Jobs***
The pie chart shows the proportion of remote, hybrid, and on-site roles.

**Insight:**  
- A large share of roles are fully remote, reflecting the growing trend in the data industry.
- Hybrid roles are also becoming common, offering flexibility.
***Note : if remote_ratio is 0 = on-site, 100 = fully remote, 50 = hybrid.***

```python
# Salary distribution
plt.figure(figsize=(8,4))
sns.histplot(df['salary_in_usd'], kde=True, color='purple')
plt.title('Salary Distribution (USD)')
plt.xlabel('Salary in USD')
plt.show()
```

# ***Salary Distribution***
This histogram shows how salaries are spread across all data science roles.

**Insight:**  
- Most salaries are clustered between $60,000 and $150,000.
- There are some very high-paying outliers above $250,000.


```python
# Average salary by experience level
exp_salary = df.groupby('experience_level')['salary_in_usd'].mean()
plt.figure(figsize=(6,4))
exp_salary.plot(kind='bar', color='teal')
plt.title('Average Salary by Experience Level')
plt.ylabel('Salary in USD')
plt.xlabel('Experience Level')
plt.show()
```

# ***Average Salary by Experience Level***
This bar chart shows how average salary changes by experience level.

**Insight:**  
- Senior and Executive-level roles earn almost 2 _ 3x more than junior roles.
- This highlights the value of experience in data science careers.
