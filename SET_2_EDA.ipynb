```python
# 2. DATA VISUALIZATION
plt.figure(figsize=(16, 4))
plt.subplot(1, 3, 1)
sns.histplot(df['Age'], kde=True, bins=20)
plt.title("Age Distribution")
plt.subplot(1, 3, 2)
sns.histplot(df['Income'], kde=True, bins=20)
plt.title("Income Distribution")
```

```python
plt.subplot(1, 3, 3)
sns.histplot(df['Spending_Score'], kde=True, bins=20)
plt.title("Spending Score Distribution")
plt.tight_layout()
plt.show()
```

```python
#Categorical Analysis
plt.figure(figsize=(10, 4))
plt.subplot(1, 2, 1)
sns.countplot(data=df, x='Gender', hue='Purchased')
plt.title("Gender vs Purchased")
```

```python
plt.subplot(1, 2, 2)
sns.countplot(data=df, x='City', hue='Loyalty_Level')
plt.title("Loyalty Level across Cities")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
#Relationship Analysis
plt.figure(figsize=(6, 5))
sns.scatterplot(data=df, x='Income', y='Spending_Score', hue='Purchased', palette='coolwarm')
plt.title("Income vs Spending Score by Purchase Decision")
plt.show()
```

```python
plt.figure(figsize=(6, 5))
sns.boxplot(data=df, x='Loyalty_Level', y='Income', palette='Set3')
plt.title("Income across Loyalty Levels")
plt.show()
```

```python
#Correlation Analysis
plt.figure(figsize=(6, 4))
corr = df.select_dtypes(include=[np.number]).corr()
sns.heatmap(corr, annot=True, cmap='Blues')
plt.title("Correlation Heatmap")
plt.show()
```
