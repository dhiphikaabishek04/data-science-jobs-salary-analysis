```python
# Function to treat outliers using IQR method
def treat_outliers_iqr(df, col):
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    print(f"\n{col}: Lower={lower:.2f}, Upper={upper:.2f}")
    # Count how many outliers
    outliers = df[(df[col] < lower) | (df[col] > upper)]
    print(f"Outliers detected in {col}: {outliers.shape[0]}")

    # Cap the outliers (winsorization)
    df[col] = np.where(df[col] < lower, lower,
              np.where(df[col] > upper, upper, df[col]))
    return df

# Apply outlier treatment to all numeric columns (except target)
for col in df.columns:
    if col != "MEDV":  # Keep target unchanged for now
        df = treat_outliers_iqr(df, col)

# Verify after treatment with boxplot
plt.figure(figsize=(12,6))
sns.boxplot(data=df, orient="h", palette="Set2")
plt.title("Boxplot After Outlier Treatment")
plt.show()
```

```python
from scipy import stats

# Copy original dataset for z-score method
df_z = df.copy()

# Calculate Z-scores for numeric columns (excluding target)
z_scores = np.abs(stats.zscore(df_z.drop("medv", axis=1)))

# Identify rows where any column has z > 3
outlier_rows = (z_scores > 3).any(axis=1)
print("Number of outliers (Z-score > 3):", outlier_rows.sum())

# Option 1: Remove outliers
df_removed = df_z[~outlier_rows]

print("Shape before removal:", df_z.shape)
print("Shape after removal:", df_removed.shape)

# Option 2: Just cap values instead of removing (Winsorization)
# Loop through each column
for col in df_z.columns:
    if col != "medv":
        mean = df_z[col].mean()
        std = df_z[col].std()
        upper = mean + 3*std
        lower = mean - 3*std
        df_z[col] = np.where(df_z[col] > upper, upper,
                   np.where(df_z[col] < lower, lower, df_z[col]))

# Verify with boxplot (after capping)
plt.figure(figsize=(12,6))
sns.boxplot(data=df_z, orient="h", palette="Set1")
plt.title("Boxplot After Z-score Outlier Treatment")
plt.show()
```

#IQR Method (capping winsorization)
Because:
The dataset has skewed features (e.g. CRIM, LSTAT).
IQR does not assume normal distribution (unlike Z-score).
Keeps dataset size intact (important for better model generalization).
Usually improves accuracy for Linear Regression & Tree-based models.
